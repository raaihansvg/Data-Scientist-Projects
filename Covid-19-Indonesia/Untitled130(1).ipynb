**Data Cleaning**

*Handle Missing Values*

```python
null = data.isnull().sum()
print(null)
```

```python
data = data.drop(['City or Regency', 'Special Status'], axis=1)
```

City or Regency column has a missing values in all rows of data and Special Status has a second most missing values and also has uninformative values

The Island, Time Zone, and Province columns have the same number of missing values ​​due to an anomalous value in the Location column, namely 'Indonesia', which should not be in the Location column

```python
data = data[data['Location'] != 'Indonesia'].reset_index(drop=True)
```

```python
CountMissingValuesRn = data.isnull().sum()
CountAllRows = data.shape

print(CountMissingValuesRn)
print("CountAllRows:", CountAllRows)
```

```python
ColTotal = ['Total Urban Villages',
            'Total Rural Villages',
            'Total Cities']

for col in ColTotal:
  data[col] = data.groupby('Location')[col].transform(lambda y: y.fillna(y.max()))
```

```python
NewCasesMedian = data['Growth Factor of New Cases'].median()
data['Growth Factor of New Cases'] = data['Growth Factor of New Cases'].fillna(NewCasesMedian)
```

```python
NewDeathsMedian = data['Growth Factor of New Deaths'].median()
data['Growth Factor of New Deaths'] = data['Growth Factor of New Deaths'].fillna(NewDeathsMedian)
```

```python
data.isnull().sum()
```

```python
print(data[data['Total Cities'].isnull()]['Location'].unique())
```

```python
data.loc[data['Location'] == 'Sulawesi Barat', 'Total Cities'] = 0
```

```python
print(data[data['Total Urban Villages'].isnull()]['Location'].unique())
```

```python
data.loc[data['Location'] == 'Aceh', 'Total Urban Villages'] = 0
```

```python
print(data[data['Total Rural Villages'].isnull()]['Location'].unique())
```

```python
data.loc[data['Location'] == 'DKI Jakarta', 'Total Rural Villages'] = 0
```

*remove column that will cz data leakage*

```python
data = data.drop('Location ISO Code', axis=1)
```

*handle data type*

```python
data.dtypes
```

The data column must be changed to datetime, the Case Fatality Rate and Case Recovered Rate columns must be changed to float and the percentage signs removed, and the Total Cities, Total Urban Villages, and Total Rural Villages columns must be changed to integer

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Case Fatality Rate'] = data['Case Fatality Rate'].str.replace('%', '').astype(float)
data['Case Recovered Rate'] = data['Case Recovered Rate'].str.replace('%', '').astype(float)
```

```python
ColumnTotal = ['Total Cities', 'Total Urban Villages',
               'Total Rural Villages']

for col in ColumnTotal:
  data[col] = data[col].astype(int)
```

```python
data['Time Zone'] = data['Time Zone'].str.strip()
```

```python
timezone_map = {
    'UTC+07:00': 0,
    'UTC+08:00': 1,
    'UTC+09:00': 2
}


data['Time Zone'] = data['Time Zone'].map(timezone_map)
```
