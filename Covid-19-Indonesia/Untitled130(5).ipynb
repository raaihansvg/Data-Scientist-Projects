**Clustering**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

last = data.sort_values('Date').groupby(
    [col for col in data.columns if col.startswith('Location_')]
).tail(1)

location_cols = [
    col for col in data.columns if col.startswith('Location')
]

last['Province'] = last[location_cols].idxmax(axis=1)
last['Province'] = last['Province'].str.replace("Location_", "")

features = last[[
    'Total Cases per Million',
    'Total Deaths per Million',
    'Case Fatality Rate',
    'Population Density'
]]


scaler = StandardScaler()
scaled_feat = scaler.fit_transform(features)

inertia = []

for x in range (1,10):
  kmeans = KMeans(n_clusters = x, random_state=42)
  kmeans.fit(scaled_feat)
  inertia.append(kmeans.inertia_)

plt.plot(range(1,10), inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
last['Cluster'] = kmeans.fit_predict(scaled_feat)
last[['Province','Cluster']].sort_values('Cluster')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_feat = pca.fit_transform(scaled_feat)

last['PC1'] = pca_feat[:,0]
last['PC2'] = pca_feat[:,1]

plt.figure(figsize=(10,6))
plt.scatter(
    last['PC1'],
    last['PerWeek'],
    c = last['Cluster']

)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering of Indonesian Provinces (PCA Projection)')
plt.show()
```

Kesimpulan:

- Cluster 3 -> DKI Jakarta. Wilayah paling padat dengan jumlah kasus tertinggi di Indonesia.

- Cluster 2 -> Jateng, Jatim, dkk,Mempunyai angka kematian yang menonjol dan butuh penguatan kapasitas rumah sakit

- Cluster 0 -> Bali, DIY, dkk, Mempunyai angka kematian terkendali yang disebabkan oleh rawannya lonjakan karena pariwisata/industri.

- Cluster 1 -> Jabar, Banten, & mayoritas provinsi lain. Kondisi paling mempunyai angka kematian terkecil karena populasi penduduk tidak terlalu besar

**Stastical Analysis**

```python
data[['Population Density', 'Total Cases per Million']].corr()
```

```python
from sklearn.linear_model import LinearRegression

X = data[['Population Density']]
y = data['Total Cases per Million']

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.xlabel('Population Density')
plt.ylabel('Total Cases per Million')
plt.title('Linear Regression: Population Density vs Cases per Million')
plt.show()
```

```python
r_squared = model.score(X, y)
print("R²:", r_squared)
```

Based on the regression results, the R2 value of 0.206 indicates that population density only has a 20.6% impact on the total number of COVID-19 cases in Indonesia.This demonstrates that non geographic factors such as community behavior (health protocols) n policy interventions play a much larger role (79.4%) in controlling the spread of the virus
