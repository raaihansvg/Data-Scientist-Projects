**JOIN DATA**

```python
orders_customer = orders.merge(customers,on="customer_id",how="left")
```

```python
data = orders_customer.merge(order_items,on="order_id",how="left")
```

**USER LEVEL DATASET**

```python
data["order_purchase_timestamp"] = pd.to_datetime(
    data["order_purchase_timestamp"]
)
```

```python
reference_date = data["order_purchase_timestamp"].max()
reference_date
```

```python
user_level = (
    data
    .groupby("customer_unique_id")
    .agg(
        total_spending=("price", "sum"),
        total_items=("order_item_id", "count"),
        total_orders=("order_id", "nunique"),
        avg_item_price=("price", "mean"),
        first_order_date=("order_purchase_timestamp", "min"),
        last_order_date=("order_purchase_timestamp", "max"),
        active_days=("order_purchase_timestamp", lambda x: x.dt.date.nunique())
    )
    .reset_index()
)
```

```python
user_level["recency_days"] = (
    reference_date - user_level["last_order_date"]
).dt.days
```

```python
user_level["avg_order_value"] = (
    user_level["total_spending"] / user_level["total_orders"]
)
```

```python
user_level = user_level.drop(
    columns=["first_order_date", "last_order_date"]
)
```

**MAIN DATA**

```python
data = user_level
data.head()
```

```python
drop_column = ['customer_unique_id','log_total_spending','total_spending',
               'avg_item_price','avg_order_value']
```

drop kolom pada 'avg_item_price','avg_order_value','total_spending' dilakukan karena ketiga kolom tersebut jika digunakan sebagai fitur akan menyebabkan data leakage, pengujian ini dilakukan sebelumnya dengan melihat korelasi fitur tersebut dengan target yang menghasilkan nilai yang sangat tinggi di antara ketiga nya

--------------------------------------------------------------------------------
Dropping the columns on 'avg_item_price', 'avg_order_value', 'total_spending' was done because if these three columns were used as features, they would cause data leakage. This test was done beforehand by looking at the correlation of these features with the target which produced very high values among the three

**SPLIT DATA**

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns = drop_column)
y = data.log_total_spending

TrainX,ValX,TrainY,ValY = train_test_split(X,y,
                                           test_size=0.3,
                                           random_state=42)


features = TrainX.columns

TrainX.shape,ValX.shape
```

**PREPROCESSING**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

transform = Pipeline(steps=[
    ('scaler',StandardScaler())
])

preprocessing = ColumnTransformer(
    transformers= [
        ('num', transform, features)
    ]
)

TrainXPrep = preprocessing.fit_transform(TrainX)
```

**MUTUAL INFORMATION**


```python
from sklearn.feature_selection import mutual_info_regression

feature_name = list(features)

mi_scores = mutual_info_regression(
    TrainXPrep,TrainY,random_state=42
)

mi_series = pd.Series(mi_scores,index=feature_name)
mi_series = mi_series.sort_values(ascending=False)

mi_series
```

```python
plt.figure(figsize=(10,6))
mi_series.head(4).plot(kind='barh')
plt.title("top 4 mutual information before modelling")
plt.show()
```

**BASELINE MODEL**

```python
from sklearn.linear_model import LinearRegression

linreg_model = Pipeline(steps=[
    ('preprocessor', preprocessing),
    ('model', LinearRegression())
])
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

rmse_score = -cross_val_score(
    linreg_model,TrainX,TrainY,
    scoring = 'neg_root_mean_squared_error',
    cv = 5
)

print("RMSE Linear Regression")
print("Mean:", rmse_score.mean())
print("Std:", rmse_score.std())
```

**RANDOM FOREST MODEL**

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = Pipeline(steps=[
    ('preprocessor', preprocessing),
    ('model', RandomForestRegressor(
        n_estimators=200, random_state=42, n_jobs=-1
    ))
])
```

```python
rf_rmse = -cross_val_score(
    rf_model,TrainX,TrainY,
    scoring='neg_root_mean_squared_error',
    cv=5
)

print("RMSE Random Forest")
print("Mean", rf_rmse.mean())
print("Std", rf_rmse.std())
```

Dengan hanya menggunakan fitur perilaku, model ini mencapai daya prediksi yang moderat. Random Forest sedikit lebih unggul daripada Regresi Linier, yang menunjukkan adanya hubungan non-linier. Namun, kesalahan secara keseluruhan menunjukkan bahwa pengeluaran pengguna sulit diprediksi tanpa sinyal nilai transaksional

--------------------------------------------------------------------------------
Using only behavioral features, the model achieves moderate predictive power. Random Forest slightly outperforms Linear Regression, indicating a non-linear relationship. However, the overall error indicates that user spending is difficult to predict without transactional value signals
