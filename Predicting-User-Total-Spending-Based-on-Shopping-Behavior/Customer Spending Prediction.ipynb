**JOIN DATA**

```python
orders_customer = orders.merge(customers,on="customer_id",how="left")
```

```python
data = orders_customer.merge(order_items,on="order_id",how="left")
```

**USER LEVEL DATASET**

```python
data["order_purchase_timestamp"] = pd.to_datetime(
    data["order_purchase_timestamp"]
)
```

```python
reference_date = data["order_purchase_timestamp"].max()
reference_date
```

```python
user_level = (
    data
    .groupby("customer_unique_id")
    .agg(
        total_spending=("price", "sum"),
        total_items=("order_item_id", "count"),
        total_orders=("order_id", "nunique"),
        avg_item_price=("price", "mean"),
        first_order_date=("order_purchase_timestamp", "min"),
        last_order_date=("order_purchase_timestamp", "max"),
        active_days=("order_purchase_timestamp", lambda x: x.dt.date.nunique())
    )
    .reset_index()
)
```

```python
user_level["recency_days"] = (
    reference_date - user_level["last_order_date"]
).dt.days
```

```python
user_level["avg_order_value"] = (
    user_level["total_spending"] / user_level["total_orders"]
)
```

```python
user_level = user_level.drop(
    columns=["first_order_date", "last_order_date"]
)
```

**MAIN DATA**

```python
data = user_level
data.head()
```
