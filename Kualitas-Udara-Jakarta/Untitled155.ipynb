```python
import pandas as pd
import glob

path = '/content/drive/MyDrive/Dataset Arkavidia/ISPU/'

files = glob.glob(path + "indeks-standar-*.csv")
simpan = []

for i in files:
  data = pd.read_csv(i)
  mapping = {
      'pm_10': 'pm10',
      'pm_duakomalima': 'pm25',
      'lokasi_spku': 'stasiun',
      'categori': 'category'
  }

  data.rename(columns=mapping, inplace=True)

  kolom_target = ['periode_data', 'tanggal', 'stasiun', 'pm10', 'pm25',
                  'so2', 'co', 'o3', 'no2', 'max', 'critical', 'category']


  simpan_kolom = [c for c in kolom_target if c in data.columns]
  data = data[simpan_kolom]

  simpan.append(data)


data_ispu_10_22 = pd.concat(simpan, axis=0, ignore_index=True)
data_ispu_10_22['tanggal'] = pd.to_datetime(data_ispu_10_22['tanggal'], errors='coerce')
data_ispu_10_22 = data_ispu_10_22.sort_values(by='tanggal').reset_index(drop=True)


print(f'Selesai GnG',{len(files)})
print(f"Rentang data: {data_ispu_10_22['tanggal'].min()} sampai {data_ispu_10_22['tanggal'].max()}")
```
